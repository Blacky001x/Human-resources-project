# Data preprocessing

```python
# Vidimo da su Attrition, Over18 i OverTime kategorijalne varijable (sa Yes i No vrednostima), njih kao takve model mašinskog učenja ne može prihvatiti, stoga ih transformišemo u numerički tip (0 i 1)
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
```

```python
employee_df.head()
```

```python
employee_df.isnull().sum().sum() # nema missinga
```

```python
# brišemo ono što nam neće trebati u modelu iz dataframea
employee_df.drop(["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"], axis = 1, inplace = True)
```

```python
left_df   = employee_df[employee_df['Attrition'] == 1] # podskup podataka, zaposleni koji su napustili firmu
stayed_df = employee_df[employee_df['Attrition'] == 0]  # i oni koji su ostali u firmi
```

```python
N_total = len(employee_df)

N_left = len(left_df)
percent_left = round(100.0*len(left_df)/len(employee_df), 2) # zaokruzeno na dve decimale

N_stayed = N_total - N_left
percent_stayed = 100.0 - percent_left


print("Firmu je napustilo: ", N_left, "što čini ", percent_left, "% zaposlenih.")
print("Sa druge strane, u firmi je ostalo ", N_stayed, "zaposlenih, odnosno ", percent_stayed, "%.")

# samo 237 ljudi je napustilo firmu, 1233 je ostalo! 
# Jako mali broj ljudi koji su napustili firmu
# Ovo će sigurno rezultirati slabijim performansama modela predikcije!
```

# Kreiranje testnog i trening seta

```python
employee_df.head()
```

```python
# IDEJA: imam kategorijalne podatke, koji se ne mogu "strpati" u machine learning model
# izdvojiću u zasebne data frameove čisto numeričke podatke i čisto kategorijalne podatke
# nad kategorijalnim podacima cu primeniti onehotencoder
# nakon čega ću spojiti ta dva datafrejma u jedan veliki
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
```

```python
X_cat
```

```python
X_cat.info()
```

```python
temp = (employee_df.dtypes == 'object')
object_cols = list(temp[temp].index)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
object_cols_oh = pd.DataFrame(onehotencoder.fit_transform(employee_df[object_cols]))
```

```python
object_cols_oh.index = employee_df.index
```

```python
object_cols_oh # dejta frejm sa cisto kategorijalnim podacima, nakon primene onehotencodera
```

```python
numerical_df = employee_df.drop(object_cols, axis=1)
```

```python
numerical_df.info() # cisto numerički dataframe
```

```python
# numerički df sadrži Attrition, njega izbacujemo pošto ćemo tu varijablu predviđati. Ukoliko to ne uradimo, data leakage će biti problem
numerical_df.drop('Attrition', axis=1, inplace=True)
```

```python
numerical_df.columns # sada je ok, nema više Attrition među featuresima
```

```python
X_total = pd.concat([object_cols_oh, numerical_df], axis = 1) # spajam numericki i "bivši" kategorijalni df
```

```python
X_total.head() # sada su svi podaci numeričkog tipa
```

```python
# vidimo da raspon varijabli u X_total varira, neke varijable su 0-1, neke od 0-10... želim da skaliram sve te podatke, zarad pokušaja da se maksimizuju performanse modela koji će biti treniran
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X_total)
```

```python
X
```

```python
y = employee_df['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

# MODEL 1: Logistička regresija

```python
X_train.shape # u train setu imam 1029 ispitanika
```

```python
X_test.shape # u test setu 441
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test))) # accuracy je poprilično dobar, ali on može biti lažno visok ako u dve kategorije nema približno isti broj ispitanika-slučajeva, a ovde nema isti broj (omnogo više ispitanika je ostalo u firmi nego što ju je napustilo)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix( y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred)) # recall je nizak! Slabo predviđamo ko će napustiti firmu
```

# MODEL 2: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test))) # slabije od logističke regresije, ali zadovoljavajuće
```

```python
cm = confusion_matrix( y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

# MODEL 3: XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
xgboost_classifier = XGBClassifier()
xgboost_classifier.fit(X_train, y_train)
y_pred = xgboost_classifier.predict(X_test)
```

```python
print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix( y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```
